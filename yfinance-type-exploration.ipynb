```python
from random import random

def download_ticker_data(ticker, period="1d", max_retries=5, base_delay=1, jitter=0.1):
  """
  Downloads historical data for a single ticker with exponential backoff.

  Args:
    ticker: The ticker symbol.
    max_retries: The maximum number of retry attempts.
    base_delay: The base delay in seconds between retries.
    jitter: A random factor to add to the delay for each retry.

  Returns:
    A pandas DataFrame containing the downloaded data or None on failure.
  """

  for attempt in range(max_retries):
    try:
      # Download data with auto-adjust
      return yf.download(ticker, period=period, auto_adjust=True)
    except Exception as e:
      delay = (2**attempt + random() * jitter) * base_delay
      print(f"Error downloading {ticker}: {e}. Retrying in {delay:.2f} seconds...")
      time.sleep(delay)

  print(f"Failed to download data for {ticker} after {max_retries} attempts.")
  return None

import yfinance as yf
import time
from random import random
import pandas as pd

def download_historical_data(tickers, period="1d", max_retries=5, base_delay=1, jitter=0.1):
  """
  Downloads historical data for a list of tickers with exponential backoff.

  Args:
    tickers: A list of ticker symbols.
    period: The period for which to download data (e.g., "1d", "5d", "max").
    max_retries: The maximum number of retry attempts per ticker.
    base_delay: The base delay in seconds between retries.
    jitter: A random factor to add to the delay for each retry.

  Returns:
    A pandas DataFrame containing the downloaded data or None on failure.
  """

  data = {}
  # Do work only once per ticker.
  for ticker in list(set(tickers)):
    data[ticker] = download_ticker_data(ticker, period, max_retries, base_delay, jitter)
    print(data[ticker])
    
  if not data:
    return None

  return data

# Example usage
tickers = df['Ticker'].tolist()
latest_data = download_historical_data(tickers, period="15d")
```

```python
from talib import ATR

def calculate_atr_values(ohlcv_dict, period=14):
  """
  Calculates ATR and related values for each ticker in a dictionary of OHLCV data.

  Args:
    ohlcv_dict: A dictionary of dictionaries, where each key is a ticker symbol and
                the value is a DataFrame containing 'Open', 'High', 'Low', 'Close', 'Volume' columns.
    period: The period for ATR calculation (default: 14).

  Returns:
    A modified dictionary where each ticker's DataFrame has added columns:
      - 'ATR': Average True Range
      - 'Close_Minus_ATR_15': Close price - ATR * 1.5
      - 'Close_Plus_ATR_3': Close price + ATR * 3
  """
    
  for ticker, ticker_data in ohlcv_dict.items():
    if period <= len(ticker_data):
        pass
      
    high = ticker_data['High']
    low = ticker_data['Low']
    close = ticker_data['Close']
    # Calculate ATR for the ticker
    atr = ATR(high, low, close, timeperiod=period)
    # Add ATR and calculated values to the ticker's DataFrame
    ticker_data['ATR'] = atr
    ticker_data['Close_Minus_ATR_15'] = close - atr * 1.5
    ticker_data['Close_Plus_ATR_3'] = close + atr * 3

  return ohlcv_dict

# Example usage
latest_data = calculate_atr_values(latest_data, period=14)
# Access data for a specific ticker:
print(latest_data)


# Download latest stock data using yfinance
# tickers = df['Ticker'].tolist()
```

```python
def process_csv(csv_file):
    """
  Processes a CSV file with added 'Tags' column and outputs desired information.

  Args:
    csv_file: The path to the CSV file.

  Returns:
    A Pandas DataFrame containing the processed data.
  """

    # Read the CSV file into a Pandas DataFrame
    # df = pd.read_csv(csv_file)

    # Datetime and timedelta (in days) to now.
    df['Date'] = pd.to_datetime(df['Date'])
    df['Timedelta'] = (datetime.now() - df['Date']).dt.days
    
    # Calculate cost basis.

    # Download latest stock data using yfinance
    # tickers = df['Ticker'].tolist()

    # latest_data = yf.download(tickers, period="1d", auto_adjust=True)
    
    # Add latest price and ATR columns to DataFrame
    for ticker, ticker_data in latest_data.items():
        df.loc[df['Ticker'] == ticker, 'Latest Price'] = ticker_data.iloc[-1]['Close']
        df.loc[df['Ticker'] == ticker, 'ATR'] = ticker_data.iloc[-1]['ATR']
        df.loc[df['Ticker'] == ticker, 'Close_Minus_ATR_15'] = ticker_data.iloc[-1]['Close_Minus_ATR_15']
        df.loc[df['Ticker'] == ticker, 'Close_Plus_ATR_3'] = ticker_data.iloc[-1]['Close_Plus_ATR_3']

        # If this ticker has no trailing stop, set one.
        df.loc[df['Ticker'] == ticker, 'Action'] = df['Trailing Stop'].where(df.loc[df['Ticker'] == ticker, 'Trailing Stop'].notna(), other="Update trailing stop.")
        df.loc[df['Ticker'] == ticker, 'Trailing Stop'] = df['Trailing Stop'].mask(df['Trailing Stop'].isna(), ticker_data.iloc[-1]['Close_Minus_ATR_15'])
    
    # df['Latest Price'] = df['Ticker'].apply(lambda t: latest_data.loc[t, 'Close'] if t in latest_data.index else None)
    
    # Calculate the price change
    df['Price Change'] = df['Latest Price'] - df['Price']
    # print(pd.DataFrame(columns=[df['Price'], df['Cost Basis'], df['Price'] - df['Cost Basis']]))
    
    # Calculate the market value
    df['Market Value'] = df['Quantity'] * df['Latest Price']

    # Calculate the total gain/loss
    df['Total Gain/Loss ($)'] = df['Market Value'] - df['Cost Basis']
    df['Total Gain/Loss (%)'] = (df['Market Value'] - df['Cost Basis']) / df['Cost Basis']
    
    # Calculate anualized rate of return.
    df['ARR'] = (df['Total Gain/Loss (%)'] / df['Timedelta']) * 365 # daily RR * 365 -> annaul RR
    
    # Calculate the percentage of account
    df['% of Acct'] = df['Market Value'] / df['Market Value'].sum() * 100

    print(df)
    
    # Group by ticker and sum the quantities, market values, etc.
    # grouped_df = df.groupby('Ticker').agg(
    #     Quantity=('Quantity', 'sum'),
    #     Price=('Price', 'mean'),
    #     Price_Change=('Price Change', 'sum'),
    #     Market_Value=('Market Value', 'sum'),
    #     Cost_Basis=('Cost Basis', 'sum'),
    #     Trailing_Stop=('Trailing Stop', 'mean'),
    #     Action=('Action', 'first'),
    #     Total_Gain_Loss_Dollars=('Total Gain/Loss ($)', 'sum'),
    #     Total_Gain_Loss_Percent=('Total Gain/Loss (%)', 'sum'),
    #     ARR=('ARR', 'mean'),
    #     Pct_of_Acct=('% of Acct', 'sum'),
    #     Timedelta=('Timedelta', 'mean'),
    #     ATR=('Timedelta', 'mean'),
    #     ATR_Low=('Close_Minus_ATR_15', 'mean'),
    #     ATR_High=('Close_Plus_ATR_3', 'mean')
    # )
    grouped_df = df.groupby('Ticker').agg(
        **{
            'Quantity': ('Quantity', 'sum'),
            'Price': ('Price', 'mean'),
            'Price Change': ('Price Change', 'mean'),
            'Latest Price': ('Latest Price', 'first'),
            'Market Value': ('Market Value', 'sum'),
            'Cost Basis': ('Cost Basis', 'sum'),
            'Trailing Stop': ('Trailing Stop', 'mean'),
            'Action': ('Action', 'first'),
            'Total Gain/Loss ($)': ('Total Gain/Loss ($)', 'sum'),
            'Total Gain/Loss (%)': ('Total Gain/Loss (%)', 'sum'),
            'ARR': ('ARR', 'mean'),
            '% of Acct': ('% of Acct', 'sum'),
            'Timedelta': ('Timedelta', 'mean'),
            'ATR': ('ATR', 'first'),
            'ATR Low': ('Close_Minus_ATR_15', 'first'),
            'ATR High': ('Close_Plus_ATR_3', 'first')
        }
    )

    print(grouped_df)
    return grouped_df


def get_portfolio_data(df, portfolio_tag):
    """
  Filters DataFrame based on specified portfolio tag.

  Args:
    df: The Pandas DataFrame containing processed data.
    portfolio_tag: The tag to filter by.

  Returns:
    A Pandas DataFrame containing filtered data.
  """

    # Split tags into lists and combine using OR condition
    filtered_df = df[df['Tags'].str.contains(portfolio_tag, na=False)]
    for tag in portfolio_tag.split(" "):
        filtered_df = filtered_df | df[df['Tags'].str.contains(tag, na=False)]

    return filtered_df


@click.command()
@click.option('--portfolio', type=str)
@click.option('--file', type=click.Path(exists=True), prompt='Enter CSV file path:')
def main(portfolio, file):
    """
  CLI interface for portfolio analysis.

  Args:
    portfolio: The tag to filter positions by.
    file: The path to the CSV file.
  """

    # Process CSV data
    df = process_csv(file)

    # Filter by portfolio tag if provided
    if portfolio:
        filtered_df = get_portfolio_data(df, portfolio)
        print(filtered_df)
    else:
        print(df)
```

```python
# Condition 1: If 'Latest Price' < 'Trailing Stop', set 'Action' to 'SELL'
out.loc[out['Latest Price'] < out['Trailing Stop'], 'Action'] = 'SELL'

# Condition 2: If 'Latest Price' > 'ATR_High', update 'Action' and 'Trailing Stop'
condition = out['Latest Price'] > out['ATR High']
out.loc[condition, 'Action'] = 'Update trailing stop'
out.loc[condition, 'Trailing Stop'] = out['ATR Low']
```
